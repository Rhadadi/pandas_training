````python
worlddata = pd.read_csv('https://github.com/QuantEcon/QuantEcon.lectures.code/raw/master/pandas_panel/countries.csv', sep=';')
worlddata.head()
````

````python
worlddata = worlddata[['Country (en)', 'Continent']]
````

````python
worlddata.head(2)
````

````python
worlddata = worlddata.rename(columns = {'Country (en)': 'Country' , 'Continent': 'Continent' })
worlddata
````

````python
realwage_f.head()
````

````python
realwage_f.transpose().head()
````

````python
merged = pd.merge(realwage_f.transpose(), worlddata,
                  how='left', left_index=True, right_on='Country')
merged.head()
````

````python
merged[merged.Continent.isnull()]
````

````python
missing_continents = {'Korea': 'Asia',
                      'Russian Federation': 'Europe',
                      'Slovak Republic': 'Europe'}


merged.Continent=merged['Continent'].fillna(merged.Country.map(missing_continents))
````

````python
merged[merged.Country=="Korea"]
````

```merged.Country.map(missing_continents)``` creates an array of NaN for non missing continent and values for missing

`fillna` receives an array and replace the missings with the values from the input array

````python
list_country = ['Central America', 'North America', 'South America']
for country in list_country:
    merged.Continent.replace(to_replace=country , value= 'Americas', inplace=True )
````

````python
merged = merged.set_index(['Continent', 'Country']).sort_index()
merged.head()
````
